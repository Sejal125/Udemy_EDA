```python
#find number of courses per subjects
data.head(2)
```

```python
data['subject'].value_counts()
```

```python
#to visualize 
sns.countplot(x='subject', data=data)
plt.xlabel("Subjects" ,fontsize=13)
plt.ylabel("Number of courses per subject",fontsize=13)
plt.xticks(rotation=65)
plt.show()
```

```python
#for which levels, udemy  courses providing the courses
data['level'].value_counts()
```

```python
sns.countplot(data['level'])
plt.xlabel("Number of courses per level" ,fontsize=13)
plt.ylabel("levels",fontsize=13)
plt.xticks(rotation=65)
plt.show()
```

```python
#to display the count of paid and free courses
data['is_paid'].value_counts()
```

```python
sns.countplot(x=data['is_paid'],    palette=['orange', 'green']
)
plt.xlabel("levels" ,fontsize=13)
plt.ylabel("Number of free and  paid courses",fontsize=13)
plt.xticks(rotation=65)
plt.show()
```

```python
#which course has more lectures(free or paid)
data.groupby('is_paid').mean(numeric_only=True)
```

```python
#which courses has higher number of subscribers free or paid?
sns.barplot(x='is_paid',y='num_subscribers',data=data)
```

```python
#which level has higher number of subscribers
sns.barplot(x='level',y='num_subscribers',data=data,palette=['orange', 'green', 'blue','red'])
plt.xticks(rotation=65)
plt.show()
```

```python
#find most popular course title
data['num_subscribers'].max()
```

```python
data[data['num_subscribers'].max()==data['num_subscribers']]['course_title']
```

```python
#display 10 most popular courses as per number of subscribers
data.sort_values(by='num_subscribers',ascending=False).head(10)[['course_title','num_subscribers']]
```

```python
top_10=data.sort_values(by='num_subscribers',ascending=False).head(10)[['course_title','num_subscribers']]
```

```python
#find the course which is having number of reviews
plt.figure(figsize=(10,4))
sns.barplot(x="subject",y="num_reviews",data=data)
```

```python
#does the price affec the number of revviews
plt.Figure(figsize=(15,6))
sns.scatterplot(x='price',y='num_reviews',data=data)
```
